```python
# Importing Libraries
import pandas as pd
```

```python
# Loading Dataset
features = pd.read_csv("feamat.csv") # Feature data stored in feamat.csv
train_data = pd.read_csv("train.csv") # training data stored in train.csv
```

```python
# Visualizing features dataframe
features.head()
```

```python
# Visualizing training dataframe
train_data.head()
```

```python
# Splitting 'Id' column into 'Chemical_Id' and 'Assay_Id'
train_data['Chemical_Id'] = train_data['Id'].apply(lambda x: str(x.split(";")[0])) 
# Using Split function to remove delimiter ';' and apply function to split data in all rows
train_data['Assay_Id'] = train_data['Id'].apply(lambda x: int(x.split(";")[1]))
features.rename(columns = {'V1':'Chemical_Id'}, inplace=True)
train_data.head()
```

```python
# Checking how many rows of a particular Assay are available
train_data['Assay_Id'].value_counts()
```

```python
# Mapping training features from features dataframe with training dataframe
mapped_features_train = train_data.merge(features, on=['Chemical_Id'])
# Generating csv of mapped training data
mapped_features_train.to_csv('mapped_features_train.csv',index=False)
mapped_features_train.head()
```

```python
# Importing test dataset
test_data = pd.read_csv("test.csv")
# Splitting 'Id' column into 'Chemical_Id' and 'Assay_Id'
test_data['Chemical_Id'] = test_data['x'].apply(lambda x: str(x.split(";")[0]))
# Using Split function to remove delimiter ';' and apply function to split data in all rows
test_data['Assay_Id'] = test_data['x'].apply(lambda x: int(x.split(";")[1]))
test_data.head()
```

```python
# Mapping testingfeatures from features dataframe with testing dataframe
mapped_features_test = test_data.merge(features, on=['Chemical_Id'])
# Generating csv of mapped testing data
mapped_features_test.to_csv('mapped_features_test.csv',index=False)
mapped_features_test.head()
```
